You can access IUPred2 via the REST API if you have a Uniprot accession number for the protein. Here is an example using the human NFL protein. Otherwise, a FASTA file with sequences can be uploaded to the [website](https://iupred2a.elte.hu/).

```python
accession = 'P07196' # Human NFL uniprot accession number
resp = requests.get('http://iupred2a.elte.hu/iupred2a/%s.json'%accession) 
if resp.status_code != 200:
    # This means something went wrong.
    raise ApiError('GET /tasks/ {}'.format(resp.status_code))
for key in resp.json():
    print(key) # information keys in 
json = resp.json() # json file with data
```

```python
plt.plot(json['iupred2']) # plot disorder probability
plt.xlabel('Residue #')
plt.ylabel('Disorder Probability')
```
